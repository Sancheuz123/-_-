```python
#8
a = np.array([[3, -5, 3], [1, 2, 1], [2, 7, -1]])
b = np.array([1, 4, 8])

def kramer(a, b):
    det_a = np.linalg.det(a)
    if det_a != 0:
        solutions = []
        for i in range(len(a)):
            modified_a = np.copy(a)
            modified_a[:, i] = b
            solutions.append(round(np.linalg.det(modified_a) / det_a, 1))

        print('Метод Крамера:')
        print('x =', solutions)

    else:
        print('Дорівнює нулю')
    return solutions
def gaussian_elimination(a, b):
    n = len(b)
    x = np.zeros(n)
    for k in range(n-1):
        for i in range(k+1, n):
            factor = a[i,k] / a[k,k]
            for j in range(k, n):
                a[i,j] -= factor * a[k,j]
            b[i] -= factor * b[k]
    x[n-1] = b[n-1] / a[n-1, n-1]
    for i in range(n-2, -1, -1):
        x[i] = (b[i] - np.dot(a[i,i+1:], x[i+1:])) / a[i,i]
    print('Метод Гауса:')
    print('x =', x.round(1))
def matrix_method(a, b):
    x = np.linalg.inv(a).dot(b)
    print('Матричний метод:')
    print('x =', x.round(1))
kramer(a, b)
gaussian_elimination(a, b)
matrix_method(a, b)
```

```python
x = np.linalg.solve(a, b)
print('Результат перевірки:')
print('x =', x.round(1))
```
